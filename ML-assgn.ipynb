```python
# Import necessary libraries
# Seaborn and Matplotlib are popular libraries for data visualization in Python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Visualizing the distribution of 'Age (day)'
# Histograms are useful for understanding the frequency distribution of numerical data.
# Here, we analyze the distribution of the 'Age (day)' column to see how the ages of concrete samples are distributed.
sns.histplot(df['Age (day)'], kde=True)  # kde=True adds a smooth curve over the histogram
plt.title("Distribution of Age (days)")  # Add a title for context
plt.xlabel("Age (days)")  # Label the x-axis to show what is being measured
plt.ylabel("Frequency")  # Label the y-axis to show the count of samples
plt.show()

# 2. Visualizing the distribution of 'Concrete compressive strength(MPa, megapascals)'
# Concrete compressive strength is a key property in determining the durability of concrete.
# A histogram helps us understand how the strength values are distributed.
sns.histplot(df['Concrete compressive strength(MPa, megapascals) '], kde=True)
plt.title("Distribution of Concrete Compressive Strength")
plt.xlabel("Strength (MPa)")  # Megapascals (MPa) are the units of strength
plt.ylabel("Frequency")
plt.show()

# 3. Pairwise Relationships Between Variables
# Pairplots allow us to visualize the relationships between all numerical variables in the dataset.
# Each scatterplot shows how two variables are related, while histograms show individual distributions.
sns.pairplot(df, diag_kind='kde')  # diag_kind='kde' adds a smoother representation for the diagonal
plt.suptitle("Pairwise Relationships Between Variables", y=1.02)  # Title for the entire plot
plt.show()

# 4. Correlation Heatmap
# A correlation matrix shows how strongly numerical variables are related to each other.
# Values close to 1 or -1 indicate a strong positive or negative correlation, respectively.
corr = df.corr()  # Compute the correlation matrix for numerical columns

# Visualizing the correlation matrix with a heatmap
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")  # fmt=".2f" ensures numbers have two decimals
plt.title("Correlation Heatmap")  # The title explains what the plot represents
plt.show()

# 5. Concrete Strength vs. Age (Scatter Plot)
# Scatter plots are effective for showing how one variable (Age) affects another (Strength).
# In this case, we analyze how the strength of concrete develops as it ages.
sns.scatterplot(x=df['Age (day)'], y=df['Concrete compressive strength(MPa, megapascals) '])
plt.title("Concrete Strength vs. Age")  # Title summarizes the key relationship being visualized
plt.xlabel("Age (days)")  # Age of concrete samples in days
plt.ylabel("Strength (MPa)")  # Compressive strength of concrete in megapascals
plt.grid(True)  # Adds a grid to make it easier to read values
plt.show()

# 6. Additional Exploration Idea: Comparing Two Key Variables
# You can analyze how different components like Cement affect Concrete Strength.
# Example: Scatter plot to see if Cement (component 1) affects compressive strength.
sns.scatterplot(x=df['Cement (component 1)(kg in a m^3 mixture)'], 
                y=df['Concrete compressive strength(MPa, megapascals) '])
plt.title("Concrete Strength vs. Cement Quantity")
plt.xlabel("Cement Quantity (kg/m^3)")
plt.ylabel("Strength (MPa)")
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define features and target
features = df.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
target = df['Concrete compressive strength(MPa, megapascals) ']

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize the Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Model Performance:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

```python
# Save the model to a file

import joblib
joblib.dump(model, 'concrete_strength_model.pkl')
print("Model saved successfully!")
```

```python
# Improve Model Performance
# Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and model
print(f"Best Parameters: {grid_search.best_params_}")
best_model = grid_search.best_estimator_

# Evaluate the optimized model
y_pred_optimized = best_model.predict(X_test)
print(f"Optimized Model R² Score: {r2_score(y_test, y_pred_optimized):.2f}")
```
